```python
import pandas as pd
df = pd.read_excel('./서울지역 대학교 위치.xlsx')
```

```python
df
```

```python
# 서울 지도 생성
seoul_map = folium.Map(location=[37.55, 126.98], tiles='Stamen Terrain', zoom_start=12)

# 대학교 위치 정보를 Marker로 표시
for name, lat, lng in zip(df["Unnamed: 0"], df.위도, df.경도) :
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)
```

```python
seoul_map
```

```python
# CircleMarker로 표시
for name, lat, lng in zip(df["Unnamed: 0"], df.위도, df.경도) :
    folium.CircleMarker([lat, lng], radius=10,
                        color='brown', fill=True, fill_color='coral',
                        fill_opacity = 0.7,
                        popup=name
                       ).add_to(seoul_map)
    
seoul_map.save('./seoul_colleges2.html')
```

```python
import json
```

```python
# 2007년에서 2017년까지 인구분포 저장
for i in range(2007, 2018):
    with open('./경기도행정구역경계.json', encoding='utf-8') as path:
        gdata = json.load(path)
        g_map = folium.Map(location=[37.5502,126.982], tiles='Stamen Terrain', zoom_start=8)
        year = str(i)
        folium.Choropleth(geo_data=gdata, data = df[year],
                          columns = [df.index, df[year]],
                          bins=[10000, 100000, 300000, 500000, 700000],
                          key_on='feature.properties.name').add_to(g_map)
        g_map.save('./gyonggi_population_' + year + '.html')
```
