```python
# 2007년에서 2017년까지 인구분포 저장
for i in range(2007, 2018):
    path = './경기도행정구역경계.json'
    try:
        gdata = json.load(open(path), encoding='utf8')
    except:
        gdata = json.load(open(path), encoding='utf-8-sig')
    g_map = folium.Map(location=[37.5502,126.982], tiles='Stamen Terrain', zoom_start=8)
    year= str(i)
    folium.Choropleth(geo_data=gdata, data = df[year],
                      columns = [df.index, df[year]],
                      bins=[10000, 100000, 300000, 500000, 700000],
                      key_on='feature.properties.name').add_to(g_map)
    g_map.save('./gyonggi_population_' + year + '.html')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
df = pd.read_csv('./auto-mpg.csv', header=None)
df.columns = sns.load_dataset('mpg').columns

df.horsepower.replace('?', np.nan, inplace=True)
df.horsepower = df.horsepower.astype('float64')
df.origin.replace({1:'USA', 2:'EU', 3:'JAPAN'}, inplace=True)
df.origin = df.origin.astype('category')
df.dropna(subset='horsepower',inplace=True)
# df.loc[df.horsepower.isna(), ‘horsepower’] = df.horsepower.mean()

count, bin_dividers = np.histogram(df.horsepower, bins=3)
```
