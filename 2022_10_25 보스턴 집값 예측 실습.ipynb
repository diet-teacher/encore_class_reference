```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
boston  = pd.read_csv('./boston_data.csv')
boston_test = pd.read_csv('./boston_test_data.csv')
X = boston.drop('medv', axis=1)
y = boston.medv
poly = PolynomialFeatures(degree=5)
X_poly = poly.fit_transform(X)
poly_test = PolynomialFeatures(degree=5)
test_poly = poly_test.fit_transform(boston_test)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=19)
from sklearn.linear_model import ElasticNet
for alpha in range(0,10):
    eln = ElasticNet(alpha=alpha*0.01, l1_ratio=0.5, normalize=False,random_state=19)
    eln.fit(X_train, y_train)
    pred_eln = eln.predict(X_test)
    print(alpha*0.01)
    print(f'mse = {mean_squared_error(y_test, pred_eln)} \n')
```

```python
# degree=3 l1_ratio=0.3 - 6400 / mse = 26.80103737960379
# degree=3 l1_ratio=0.5 - 3800 / mse = 26.804859664496256
# degree=3 l1_ratio=0.9 - 2100 / mse = 26.806170374670266


# degree=4 l1_ratio=0.5 - 904000 / mse = 25.55947347898077
```

```python
# PolynomialFeatures(degree=3)

eln = ElasticNet(alpha=3800, l1_ratio=0.5, normalize=False)
eln.fit(X_poly, y)
pred_eln = eln.predict(test_poly)
bos = pd.read_csv('./bos_test3.csv')
bosid  = bos.iloc[:,:1]
bosid['price'] = pred_eln
bosid.set_index('id', inplace=True)
bosid.to_csv('boston_eln_answer2.csv')
```

```python
# PolynomialFeatures(degree=4)

eln = ElasticNet(alpha=904000, l1_ratio=0.5, normalize=False)
eln.fit(X_poly, y)
pred_eln = eln.predict(test_poly)
bos = pd.read_csv('./bos_test3.csv')
bosid  = bos.iloc[:,:1]
bosid['price'] = pred_eln
bosid.set_index('id', inplace=True)
bosid.to_csv('boston_eln_answer1.csv')
```

```python
from hyperopt import fmin, hp, tpe, Trials, space_eval, STATUS_OK

def eln_objective(search_space):
    model = ElasticNet(**search_space)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    return {'loss': mse, 'status': STATUS_OK}

# new search space
search_space={'alpha':hp.choice('alpha', range(0,1000,10)),
              'l1_ratio':hp.uniform('l1_ratio', 0.1, 1),
              'max_iter':hp.choice('max_iter', range(1, 30)),
              'tol':hp.uniform('tol', 0.0001, 0.001)}


# set the hyperparam tuning algorithm
algorithm=tpe.suggest
# implement Hyperopt
best_params = fmin(fn=eln_objective,
                   space=search_space,
                   algo=algorithm,
                   max_evals=500)

params1 = space_eval(search_space, best_params)
```
