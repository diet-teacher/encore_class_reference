```python
def xml_to_yolo_bbox(bbox, w, h):
    # xmin, ymin, xmax, ymax
    x_center = ((bbox[2] + bbox[0]) / 2) / w
    y_center = ((bbox[3] + bbox[1]) / 2) / h
    width = (bbox[2] - bbox[0]) / w
    height = (bbox[3] - bbox[1]) / h
    return [x_center, y_center, width, height]
def yolo_to_xml_bbox(bbox, w, h):
    # x_center, y_center width heigth
    w_half_len = (bbox[2] * w) / 2
    h_half_len = (bbox[3] * h) / 2
    xmin = int((bbox[0] * w) - w_half_len)
    ymin = int((bbox[1] * h) - h_half_len)
    xmax = int((bbox[0] * w) + w_half_len)
    ymax = int((bbox[1] * h) + h_half_len)
    return [xmin, ymin, xmax, ymax]
```

```python
# XML to TXT
import os
import glob
import pandas as pd
import io
import xml.etree.ElementTree as ET
import argparse
def xml_to_txt(path):
    
    for xml_file in glob.glob(path + '/*.xml'):
        tree = ET.parse(xml_file)
        root = tree.getroot()
        filename = root.find('filename').text
        width = int(root.find('size').find('width').text)
        height = int(root.find('size').find('height').text)
        for member in root.findall('object'):
            lst = []
            bndbox = member.find('bndbox')
            value = (member.find('name').text,
                    (int(bndbox.find('xmax').text) - int(bndbox.find('xmin').text))/width,
                    (int(bndbox.find('ymax').text) - int(bndbox.find('ymin').text))/height,
                    ((int(bndbox.find('xmax').text) + int(bndbox.find('xmin').text))/2)/width,
                    ((int(bndbox.find('ymax').text) + int(bndbox.find('ymin').text))/2))/height
            to_txt = open(xml_file[:-4]+".txt", 'w')
            for i in value:
                to_txt.write(str(i)+' ')
            to_txt.close()
    return
```
