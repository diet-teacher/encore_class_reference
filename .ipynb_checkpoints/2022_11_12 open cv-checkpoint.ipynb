```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


img = np.zeros(shape=(512,512), dtype= np.uint8) +255

pt1 = 200, 200 # 첫번째 점의 좌표 생성
pt2 = 300, 300 # 두번째 점의 좌표 생성

src = cv2.rectangle(img, pt1, pt2, (0,0,0), -1)

gx = cv2.Sobel(src, -1, 1, 0, ksize=3) # ddepth가 -1이면 알아서 자동으로 맞춰줌 값이 -로 나오면 0으로 치환
gy = cv2.Sobel(src, -1, 0, 1, ksize=3) 

gx21 = cv2.Sobel(src, -1, 1, 0, ksize=21)
gy21 = cv2.Sobel(src, -1, 0, 1, ksize=21)

gxdd = cv2.Sobel(src, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
gydd = cv2.Sobel(src, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)

gxdda = np.abs(gxdd) # 원본값을 절대값을 해줌
gydda = np.abs(gydd)  # 원본값을 절대값을 해줌

add = gxdda + gydda # 가로 세로 더해줌

gxo = np.sqrt(gxdda)# 루트 값을 해줌
gyo = np.sqrt(gxdda)# 루트 값을 해줌

print('ddepth -1')
print(gx[195:205, 195:205])
print(gx[295:305, 295:305])

print(gy[195:205, 195:205])
print(gy[295:305, 295:305])

print('ddepth=cv2.CV_32F')
print(gxdd[195:205, 195:205])
print(gxdd[295:305, 295:305])

print(gydd[195:205, 195:205])
print(gydd[295:305, 295:305])

print('ddepth=cv2.CV_32F 절대값')
print(gxdda[195:205, 195:205])
print(gxdda[295:305, 295:305])

print(gydda[195:205, 195:205])
print(gydda[295:305, 295:305])

print('ddepth=cv2.CV_32F 루트')
print(gxo[195:205, 195:205])
print(gxo[295:305, 295:305])

print(gyo[195:205, 195:205])
print(gyo[295:305, 295:305])

cv2.imshow('src', src)
cv2.imshow('gx', gx) #줄이 하나마나 나오는 이유는 마이너스는 0으로 되기 때문
cv2.imshow('gy', gy)
cv2.imshow('gx21', gx21)
cv2.imshow('gy21', gy21)
cv2.imshow('gxdda', gxdda)
cv2.imshow('gydda', gydda)
cv2.imshow('add', add)

cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# 강사님 코드

import cv2
import numpy as np
img = np.zeros(shape=(512,512), dtype=np.uint8)+255
pt1 = 200,200
pt2 = 300,300
src = cv2.rectangle(img, pt1, pt2, (0,0,0), -1)
gx = cv2.Sobel(src, ddepth=-1, dx=1, dy=0, ksize=3)
gy = cv2.Sobel(src, ddepth=-1, dx=0, dy=1, ksize=3)
gxo = cv2.Sobel(src, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
gyo = cv2.Sobel(src, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)
print('- ddepth = -1')
print(gx[195:205, 195:205])
print(gx[295:305, 295:305])
print(gy[195:205, 195:205])
print(gy[295:305, 295:305])
print('- ddepth = cv2.CV_32F')
print(gxo[195:205, 195:205])
print(gxo[295:305, 295:305])
print(gyo[195:205, 195:205])
print(gyo[295:305, 295:305])
gxo = np.abs(gxo)
gyo = np.abs(gyo)
print('- 32F after absolute')
print(gxo[195:205, 195:205])
print(gxo[295:305, 295:305])
print(gyo[195:205, 195:205])
print(gyo[295:305, 295:305])
gxo = np.sqrt(gxo)
gyo = np.sqrt(gyo)
print('- 32F after root')
print(gxo[195:205, 195:205])
print(gxo[295:305, 295:305])
print(gyo[195:205, 195:205])
print(gyo[295:305, 295:305])
gxo = cv2.normalize(gxo, 0, 255, norm_type=cv2.NORM_MINMAX)
gyo = cv2.normalize(gyo, 0, 255, norm_type=cv2.NORM_MINMAX)
print('- 32F after stretching')
print(gxo[195:205, 195:205])
print(gxo[295:305, 295:305])
print(gyo[195:205, 195:205])
print(gyo[295:305, 295:305])
cv2.imshow('src', src)
cv2.imshow('gxo', gxo)
cv2.imshow('gyo', gyo)
cv2.imshow('gxo+gyo', gxo+gyo)
cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


img = np.zeros(shape=(512,512), dtype= np.uint8) +255

pt1 = 200, 200 # 첫번째 점의 좌표 생성
pt2 = 300, 300 # 두번째 점의 좌표 생성

src = cv2.rectangle(img, pt1, pt2, (0,0,0), -1)

gx = cv2.Sobel(src, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
gy = cv2.Sobel(src, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)

mag = cv2.magnitude(gx, gy)

minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(mag)
print('mag =', minVal, maxVal, minLoc, maxLoc)
dstM = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(dstM)
print('mag_n =', minVal, maxVal, minLoc, maxLoc)

cv2.imshow('gx', gx)
cv2.imshow('gy', gy)
cv2.imshow('mag', mag)
cv2.imshow('dstM', dstM)
cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# 2차 미분 가우시안 블러 적용 후 

import cv2
import numpy as np
src = cv2.imread('./lena.jpg', 0)
dst1 = cv2.GaussianBlur(src, ksize=(3,3), sigmaX=0.0, borderType=cv2.BORDER_CONSTANT)
lap = cv2.Laplacian(dst1, cv2.CV_32F)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(lap)
print('lap :', minVal, maxVal, minLoc, maxLoc)
print(lap[:10,:10])
dst = cv2.convertScaleAbs(lap)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(dst)
print('dst :', minVal, maxVal, minLoc, maxLoc)
print(dst[:10,:10])
dst_n = cv2.normalize(dst, None, 0, 255, cv2.NORM_MINMAX)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(dst_n)
print('dst_n :', minVal, maxVal, minLoc, maxLoc)
print(dst_n[:10,:10])

cv2.imshow('lap', lap)
cv2.imshow('dst', dst)
cv2.imshow('dst_n', dst_n)
cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# 2차 미분 가우시안 블러 적용 후 

import cv2
import numpy as np
src = cv2.imread('./lena.jpg', 0)

dstg1 = cv2.GaussianBlur(src, ksize=(3,3), sigmaX=0.0, borderType=cv2.BORDER_CONSTANT)
dstg2 = cv2.GaussianBlur(src, ksize=(3,3), sigmaX=10.0, borderType=cv2.BORDER_CONSTANT)
dstg3 = cv2.GaussianBlur(src, ksize=(3,3), sigmaX=1.0, sigmaY=10.0, borderType=cv2.BORDER_CONSTANT)

lap1 = cv2.Laplacian(dstg1, cv2.CV_32F)
lap2 = cv2.Laplacian(dstg2, cv2.CV_32F)
lap3 = cv2.Laplacian(dstg3, cv2.CV_32F)


dst1 = cv2.convertScaleAbs(lap1)
dst2 = cv2.convertScaleAbs(lap2)
dst3 = cv2.convertScaleAbs(lap3)

dst_n1 = cv2.normalize(dst1, None, 0, 255, cv2.NORM_MINMAX)
dst_n2 = cv2.normalize(dst2, None, 0, 255, cv2.NORM_MINMAX)
dst_n3 = cv2.normalize(dst3, None, 0, 255, cv2.NORM_MINMAX)

gx = cv2.Sobel(src, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
gy = cv2.Sobel(src, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)
mag = cv2.magnitude(gx, gy)
dstM = cv2.normalize(mag, 0, 255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)


cv2.imshow('dstM', dstM)
cv2.imshow('dst1', dst1)
cv2.imshow('dst_n1', dst_n1)
cv2.imshow('dst_n2', dst_n2)
cv2.imshow('dst_n3', dst_n3)
cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
src = cv2.imread('/Users/iseung-won/morphology.jpg')


kernel3 = cv2.getStructuringElement(shape=cv2.MORPH_RECT, ksize=(3,3))
kernel = cv2.getStructuringElement(shape=cv2.MORPH_RECT, ksize=(9,9)) # 노이즈를 한번에 없애기에는 커널 사이즈가 작은 것 같아 (9,9)커널을 만들었음.

open1 = cv2.morphologyEx(src, cv2.MORPH_OPEN, kernel, iterations=1)
close1 = cv2.morphologyEx(src, cv2.MORPH_CLOSE, kernel, iterations=1)
open2 = cv2.morphologyEx(close1, cv2.MORPH_OPEN, kernel, iterations=1) 
close2 = cv2.morphologyEx(open1, cv2.MORPH_CLOSE, kernel, iterations=1)
tophat1 = cv2.morphologyEx(src, cv2.MORPH_TOPHAT, kernel, iterations=1)
blackhat1 = cv2.morphologyEx(src, cv2.MORPH_BLACKHAT, kernel, iterations=1)
noise = tophat1 + blackhat1
gradient1 = cv2.morphologyEx(open2, cv2.MORPH_GRADIENT, kernel3, iterations=1) #ksize=(9,9)는 너무 굵은 선으로 나와서 (3,3)으로 바꿈
# 노이즈가 없어진 open2에 (3,3) 사각형 커널 만큼 팽창을 하면 외각이 한칸씩 늘어나고 침식을 외각이 한캄씩 줄어드는데 
# gradient를 하면 팽창된 상자에서 침식된 상자를 뺴기 떄문에 노이즈 없는 모양의 외각만 볼 수가 있음

dstx = cv2.Sobel(close2, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=5)
dsty = cv2.Sobel(close2, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=5)

absx = np.abs(dstx)
absy = np.abs(dsty)

sqrx = np.sqrt(absx)
sqry = np.sqrt(absy)

xn = cv2.normalize(sqrx, None, 0, 255, cv2.NORM_MINMAX)
yn = cv2.normalize(sqry, None, 0, 255, cv2.NORM_MINMAX)

add_abs = xn + yn

cv2.imshow('src', src)
cv2.imshow('gradient1', gradient1)
cv2.imshow('open2', open2) # 노이즈만 제거 됨
# cv2.imshow('close2', close2) # 노이즈만 제거 됨. 같은 효과임
cv2.imshow('noise', noise) #노이즈만 출력
# cv2.imshow('dstx', dstx)
# cv2.imshow('dsty', dsty)
cv2.imshow('add_abs_is_solved_with_sobel', add_abs)


cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

kernel = cv2.getStructuringElement(shape=cv2.MORPH_RECT, ksize=(3,3))
# 직접 만든 영상
src = cv2.imread('./morphology.jpg', cv2.IMREAD_GRAYSCALE)
# erosion 부식, 침식 뜻, 영상을 erosion 한 결과
erode = cv2.erode(src, kernel, iterations = 5)
# 노이즈들은 제거했으나 부작용으로 홀이 커짐 (침식됨)
# dilation 확장, 팽창 뜻, 영상을 dilation 한 결과
dilate = cv2.dilate(src, kernel, iterations = 5)
# 홀은 메꾸어주나 밖에 있는 노이즈들이 커지는 부작용
# 영상을 dilation 한 후 erosion 한 결과
erode2 = cv2.erode(dilate, kernel, iterations = 7)
# 영상을 dilation 한 후 erosion 한 후 다시 dilation 한 결과
dilate2 = cv2.dilate(erode2, kernel, iterations = 2)
# iterations은 erosion과 dilation을 반복적으로 쓸 수 있게 하는 것, 한번만 하는 것이 아닌 5번 7번 반복하겠다는 의미
cv2.imshow('src', src)
cv2.imshow('erode', erode)
cv2.imshow('dilate', dilate)
cv2.imshow('erode2', erode2)
cv2.imshow('dilate2', dilate2)
cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)




src = cv2.imread('./morphology.jpg', cv2.IMREAD_GRAYSCALE)
kernel = cv2.getStructuringElement(shape=cv2.MORPH_RECT, ksize=(3,3))
closing = cv2.morphologyEx(src, cv2.MORPH_CLOSE, kernel, iterations = 5)
opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel, iterations = 5)
gradient = cv2.morphologyEx(opening, cv2.MORPH_GRADIENT, kernel)
#gradient = cv2.morphologyEx(opening, cv2.MORPH_GRADIENT, kernel, iterations = 5)
tophat = cv2.morphologyEx(src, cv2.MORPH_TOPHAT, kernel, iterations = 5)
blackhat = cv2.morphologyEx(src, cv2.MORPH_BLACKHAT, kernel, iterations = 5)
cv2.imshow('closing', closing)
cv2.imshow('opening', opening)
cv2.imshow('gradient', gradient)
cv2.imshow('tophat', tophat)
cv2.imshow('blackhat', blackhat)
cv2.waitKey()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
