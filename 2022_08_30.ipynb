```python
def sum_besu(start, end, besu):
    total = 0
    for i in range(start, end+1):
        if i % besu == 0:
            total += i
    return total
    
    
start = int(input('시작 수를 입력하세요. :'))
end = int(input('끝 수를 입력하세요. : '))
besu = int(input('합계를 구할 배수를 입력하세요. : '))
result = sum_besu(start, end, besu)
print('%d~%d의 정수 중 %d의 배수의 합 : %d'%(start, end, besu,result))
```

```python
num1 = int(input('첫 번째 수를 입력하세요.'))
num2 = int(input('두 번째 수를 입력하세요.'))

def compute_max_gong(a,b):
    li = []
    if a > b:
        for i in range(1,a+1):
            if a % i == 0 and b % i ==0:
                li.append(i)
            
    elif a < b:
        for i in range(1,b+1):
            if a % i == 0 and b % i ==0:
                li.append(i)
           
                
    li.sort()
    result = li[-1]
    return result

max_gong = compute_max_gong(num1,num2)

print('%d와 %d의 최대공약수 : %d'%(num1, num2,max_gong))
```

```python
def average(*args):
    total = 0
    avg = 0
    for i in args:
        total += i
    avg = total / len(args)
    print('%d과목의 평균 : %.2f'%(len(args), avg))

average(80,90,100)
```
